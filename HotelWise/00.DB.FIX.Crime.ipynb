```python
crime_df_content = "Crime_In_The_USA/CrimeInTheUSA.parquet"
crime_df_dataset = pd.read_parquet(crime_df_content)
```

## Combinación db Hoteles con db Seguridad

```python
# Realizar la fusión
hoteles_final = pd.merge(user_reviews_dataset, crime_df_dataset[['STATE', 'CITY', 'CRIME RATE']], 
                         left_on=[user_reviews_dataset['STATE'].str.lower(), user_reviews_dataset['CITY'].str.lower()], 
                         right_on=[crime_df_dataset['STATE'].str.lower(), crime_df_dataset['CITY'].str.lower()], 
                         how='left')

# Eliminar las columnas adicionales
hoteles_final.drop(['key_0', 'key_1', 'CITY_y', 'STATE_y'], axis=1, inplace=True)
hoteles_final.rename(columns={'CITY_x': 'CITY', 'STATE_x': 'STATE'}, inplace=True)
```

```python
hoteles_final.head(2)
```

###  Guardamos el df con las columnas necesarias para el análisis posterior

```python
hoteles_para_guardar=hoteles_final[['NAME','LATITUDE','LONGITUDE','CITY','COUNTY', 'STATE', 'TEXT','AVG_RATING','CRIME RATE', 'AMENITIES']]
```

```python
hoteles_para_guardar.rename(columns={'TEXT': 'REVIEWS','CRIME RATE':'CRIME_RATE'}, inplace=True)
```

```python
hoteles_para_guardar.head(2)
```

```python
# Guardar DataFrame en un archivo Parquet
hoteles_para_guardar.to_parquet('hoteles_unificado.parquet')
```
