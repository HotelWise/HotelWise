# Base de Datos de Seguridad

## Extracción de los Datos

### Crimen en los Estados Unidos según recopilación del FBI de delitos conocidos por las autoridades, por estado y por ciudad.

* Estadisticas FBI 2017: https://ucr.fbi.gov/crime-in-the-u.s/2017/crime-in-the-u.s.-2017 
* Tabla: https://ucr.fbi.gov/crime-in-the-u.s/2017/crime-in-the-u.s.-2017/tables/table-8/table-8.xls
* Estadisticas FBI 2018: https://ucr.fbi.gov/crime-in-the-u.s/2018/crime-in-the-u.s.-2018 
* Tabla: https://ucr.fbi.gov/crime-in-the-u.s/2018/crime-in-the-u.s.-2018/tables/table-8/table-8.xls

MONTAJE DE ARCHIVOS:

```python
# URL de los archivos Excel 2017 y 2018
url2017 = "https://ucr.fbi.gov/crime-in-the-u.s/2017/crime-in-the-u.s.-2017/tables/table-8/table-8.xls"
url2018 = "https://ucr.fbi.gov/crime-in-the-u.s/2018/crime-in-the-u.s.-2018/tables/table-8/table-8.xls"

# Nombre del archivo local
nombre_archivo2017 = "2017table-8.xls"
nombre_archivo2018 = "2018table-8.xls"

# Descargar el archivo desde la URL
response2017 = requests.get(url2017)
response2018 = requests.get(url2018)

# Guardar el contenido descargado en archivos locales
with open(nombre_archivo2017, 'wb') as f:
    f.write(response2017.content)
with open(nombre_archivo2018, 'wb') as f:
    f.write(response2018.content)

# Cargar el archivo Excel en un DataFrame
crime_df_2017 = pd.read_excel(nombre_archivo2017,skiprows=3)
crime_df_2018 = pd.read_excel(nombre_archivo2018,skiprows=3)
```

```python
os.remove(nombre_archivo2017)
os.remove(nombre_archivo2018)
```

```python
crime_df_2017.head()
```

```python
crime_df_2018.head()
```

### Formateo de Celdas

```python
# Quitar \ y \n de los nombres de las columnas
nuevos_nombres = {nombre_columna: nombre_columna.replace("\\", " ").replace("\n", " ") for nombre_columna in crime_df_2017.columns}
crime_df_2017.rename(columns=nuevos_nombres, inplace=True)
nuevos_nombres = {nombre_columna: nombre_columna.replace("\\", " ").replace("\n", " ") for nombre_columna in crime_df_2018.columns}
crime_df_2018.rename(columns=nuevos_nombres, inplace=True)
```

```python
crime_df_2017.head(0)
```

```python
crime_df_2018.head(0)
```

```python
# Convertir todos los nombres de las columnas a mayúsculas
crime_df_2017.columns = crime_df_2017.columns.str.upper()
crime_df_2018.columns = crime_df_2018.columns.str.upper()
```

### Cálculo de la tasa de delincuencia

#### 2017

```python
crime_df_2017['TOTAL CRIME'] = crime_df_2017.loc[:, 'VIOLENT CRIME':'ARSON2'].sum(axis=1)
crime_df_2017.head()
```

```python
crime_df_2017['CRIME RATE'] = (crime_df_2017['TOTAL CRIME'] / crime_df_2017['POPULATION']) * 1000
```

```python
crime_df_2017['CRIME RATE'] = crime_df_2017['CRIME RATE'].round(2)
```

```python
crime_df_2017['STATE'] = crime_df_2017['STATE'].ffill()
```

```python
crime_df_2017.head()
```

```python
len(crime_df_2017)
```

#### 2018

```python
crime_df_2018['TOTAL CRIME'] = crime_df_2018.loc[:, 'VIOLENT CRIME':'ARSON2'].sum(axis=1)
crime_df_2018.head()
```

```python
crime_df_2018['CRIME RATE'] = (crime_df_2018['TOTAL CRIME'] / crime_df_2018['POPULATION']) * 1000
```

```python
crime_df_2018['CRIME RATE'] = crime_df_2018['CRIME RATE'].round(2)
```

```python
crime_df_2018['STATE'] = crime_df_2018['STATE'].ffill()
```

### Reordenamiento de Columnas

```python
crime_df_2017 = crime_df_2017[['STATE', 'CITY', 'POPULATION', 'TOTAL CRIME','CRIME RATE']]
crime_df_2018 = crime_df_2018[['STATE', 'CITY', 'POPULATION', 'TOTAL CRIME','CRIME RATE']]
```

```python
crime_df_2017.dropna(subset=['POPULATION', 'TOTAL CRIME'], inplace=True)
crime_df_2018.dropna(subset=['POPULATION', 'TOTAL CRIME'], inplace=True)
```

### Fusión de las Bases 2017 y 2018

```python
crime_df_merged = pd.merge(crime_df_2017, crime_df_2018, on=['STATE', 'CITY'])
crime_df_merged.rename(columns={'TOTAL CRIME_x': 'TOTAL_CRIME_2017', 'TOTAL CRIME_y': 'TOTAL_CRIME_2018'}, inplace=True)
crime_df_merged.rename(columns={'CRIME RATE_x': 'CRIME_RATE_2017', 'CRIME RATE_y': 'CRIME_RATE_2018'}, inplace=True)
crime_df_merged.rename(columns={'POPULATION_x': 'POPULATION_2017', 'POPULATION_y': 'POPULATION_2018'}, inplace=True)
```

#### Creacion Indice de seguridad

* Es importante aclarar que la resta se hace al revés de lo que debería ser, ya que para la lógica implementada en el modelo de Machine Learning, el numero poitivo mas alto indica mejor indice de seguridad, por lo que una disminución en la tasa de criminalidad se verá como un número positivo, mientras que un aumento se verá como un numero negativo.

```python
crime_df_merged['CRIME_RATE'] = crime_df_merged['CRIME_RATE_2017'] - crime_df_merged['CRIME_RATE_2018']
```

```python
crime_df_merged.head()
```

### Análisis sobre la tasa de variación

#### 2017-2018

```python
top_10 = crime_df_merged.nlargest(10, 'CRIME_RATE')[['STATE', 'CITY', 'CRIME_RATE']]
bottom_10 = crime_df_merged.nsmallest(10, 'CRIME_RATE')[['STATE', 'CITY', 'CRIME_RATE']]
print("Top 10:")
print(top_10)
print("\nBottom 10:")
print(bottom_10)
```

```python
crime_df_merged.info()
```

```python
crime_df_merged['CRIME_RATE'].describe()
```

* Se puede decir que los outliers para el indice de criminalidad creado, se encuentran fuera del rango [-100, 100], por lo tanto se pueden quitar de la lista los que no cumplen con dicha condición.

```python
len(crime_df_merged)
```

```python
filtered_crime_df_merged = crime_df_merged[(crime_df_merged['CRIME_RATE'] <= 100) & (crime_df_merged['CRIME_RATE'] >= -100)]
```
